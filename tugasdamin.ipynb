```python
import pandas as pd
import numpy as np
import collections
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
import csv
from sklearn.model_selection import RepeatedKFold

# Python script for confusion matrix creation
# Skrip python untuk pembuatan matriks
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.metrics import f1_score

#Import scikit-learn metrics module for accuracy calculation
#Import modul metrik scikit-learn untuk perhitungan akurasi
from sklearn import metrics
```

```python
# membaca data pada format csv
df = pd.read_csv("datadamin.csv")
# melihat data set sebanyak 10 baris
df.head(10)
```

```python
#mengganti setiap atribut menjadi angka agar mudah untuk diolah
df['Buying'] = df['Buying'].replace({'vhigh': 1, 'high': 2, 'med': 3, 'low':4})
df['Maint'] = df['Maint'].replace({'vhigh': 1, 'high': 2, 'med': 3, 'low':4})
df['lug_boot'] = df['lug_boot'].replace({'small': 1, 'med': 2, 'big': 3})
df['Safety'] = df['Safety'].replace({'low': 1, 'med': 2, 'high': 3})
df['Class'] = df['Class'].replace({'acc': 1, 'unacc': 2, 'good': 3, 'vgood': 4})
```

```python
df
```

```python
X = df.iloc[:, :-1].values
y = df.iloc[:, 6].values
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20)
```

```python
gnb = GaussianNB()
gnb.fit(X_train, y_train)
y_pred = gnb.predict(X_test)
```

```python
y_pred
```

```python
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
print(confusion_matrix(y_test, y_pred))
```

```python
print("Classification report")
print(classification_report(y_test,y_pred))
```
